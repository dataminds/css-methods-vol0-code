`money = 1000` 은 "money 상자에 1000을 넣어라"입니다. `=` 는 "같다"가 아니라 "넣어라"입니다.
아래를 실행해 보세요. 결과가 1100 이 나옵니다.


````python
money = 1000
money = money + 100      # 지금 값(1000)에 100을 더해 다시 money에 넣는다
print(money)
````

## 3. 반복: 같은 일을 백 번 (§1.3)
100달 동안 매달 100원씩 저금하면? 100줄을 쓰지 않고 **반복**으로 시킵니다.
`range(100)` = 100번, 들여쓴 줄이 반복의 몸통입니다.


````python
money = 0
for month in range(100):
    money = money + 100
print(money)      # 10000
````
