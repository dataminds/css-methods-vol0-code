## 1. 변수: 이름을 붙인 값 (§1.2)
`money = 1000` 은 "money 상자에 1000을 넣어라"입니다. `=` 는 "같다"가 아니라 "넣어라"입니다.
아래를 실행해 보세요. 결과가 1100 이 나옵니다.


```python
money = 1000
money = money + 100      # 지금 값(1000)에 100을 더해 다시 money에 넣는다
print(money)
```

## 2. 손 실행과 맞춰 보기
먼저 **종이에** 아래 세 줄이 끝나면 money가 얼마일지 단계별로 적어 보세요(손 실행).
그다음 실행해서 여러분이 손으로 쫓아간 값과 같은지 맞춰 보세요.


```python
money = 500
money = money + 200
money = money - 300
print(money)      # 종이에 먼저 적고 나서 확인하세요
```

## 3. 반복: 같은 일을 백 번 (§1.3)
100달 동안 매달 100원씩 저금하면? 100줄을 쓰지 않고 **반복**으로 시킵니다.
`range(100)` = 100번, 들여쓴 줄이 반복의 몸통입니다.


```python
money = 0
for month in range(100):
    money = money + 100
print(money)      # 10000
```

## 4. 직접 바꿔 보기
- 3번에서 매달 저금액 `100` 을 `250` 으로, `range(100)` 을 `range(12)` 로 바꿔 보세요. 얼마가 되나요?
- 극단값으로 확인: 저금액을 `0` 으로 두면 결과가 `0` 이어야 합니다(그렇지 않으면 어딘가 틀린 것).
