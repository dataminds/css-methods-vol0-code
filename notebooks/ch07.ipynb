## 1. softmax: 점수를 선택 확률로
점수 목록을 확률 목록으로 바꾸는 변환기입니다. 손잡이 하나(온도 모수 τ, 타우)로 '단호함'을 조절합니다.
τ가 낮으면 거의 최고점만, 높으면 아무거나 고릅니다.


```python
import math

def softmax_two(v_a, v_b, tau):
    ea = math.exp(v_a / tau)
    eb = math.exp(v_b / tau)
    return ea / (ea + eb)        # A를 고를 확률
```

## 2. 온도 모수 표 재현 (§7.3)
기사 A(점수 2)와 B(점수 1). τ를 바꾸면 A를 고를 확률이 어떻게 변하는지 봅니다.
책의 서명: 52% · 62% · 73% · 88% · 98%.


```python
print('τ       A를 고를 확률   느낌')
느낌 = {10:'거의 동전 던지기', 2:'살짝 기욺', 1:'뚜렷이 기욺', 0.5:'꽤 단호', 0.25:'사실상 max()'}
for tau in (10, 2, 1, 0.5, 0.25):
    p = softmax_two(2, 1, tau)
    print(f'{tau:>5}    {p*100:>3.0f}%          {느낌[tau]}')
```

## 3. 선택지가 셋일 때 (§7.4)
세 단계(나누기·지수·비율)는 선택지가 늘어도 똑같습니다. A(2)·B(1)·C(0), τ=1.
확인: 셋을 더하면 100%인가? 점수 순서와 확률 순서가 같은가?


```python
def softmax(scores, tau):
    exps = [math.exp(v / tau) for v in scores]
    total = sum(exps)
    return [e / total for e in exps]

probs = softmax([2, 1, 0], tau=1)
print('확률:', [f'{p*100:.0f}%' for p in probs])
print('합계:', f'{sum(probs)*100:.0f}%')
```
