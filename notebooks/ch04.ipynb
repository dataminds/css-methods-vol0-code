## 2. 같은 규칙, 다른 세계 (§4.1)
같은 eps라도 씨앗(우연)만 다르면 결과가 달라집니다. 씨앗 세 개를 돌려 봅니다.


```python
from w03_bounded_confidence import BoundedConfidence, clusters

def one(seed, eps=0.15, steps=60, n=100):
    m = BoundedConfidence(n=n, eps=eps, seed=seed)
    for _ in range(steps):
        m.step()
    return clusters(m)

for s in (73, 37, 5):
    print(f'씨앗 {s:>3}: 군집 {one(s)}개')
```

## 3. 분포로 보기 (§4.2)
씨앗 20개를 돌려 '군집 수가 몇 번 나왔나'를 셉니다. 이 표가 분포입니다.
eps=0.15 에서 책의 서명은 `{2: 7, 3: 13}` 입니다.


```python
from abmlib import run_seeds
from collections import Counter

eps = 0.15            # ← 이 값을 바꿔 보세요 (예: 0.05, 0.30)
dist = run_seeds(BoundedConfidence, clusters,
                 seeds=range(1, 21), steps=60, eps=eps, n=100)
표 = dict(sorted(Counter(int(x) for x in dist).items()))
print(f'eps={eps}, 씨앗 1~20, n=100, 60단계')
print('군집 수 분포:', 표)
print(f'평균 {dist.mean():.2f} · 범위 {int(dist.min())}~{int(dist.max())}')
```

## 4. 분포 그림


```python
from abmlib import plot_hist
from IPython.display import Image, display
plot_hist([int(x) for x in dist], 'ch04_hist.png', f'eps={eps}: 군집 수 분포 (씨앗 20개)')
display(Image('ch04_hist.png'))
```

## 5. 파라미터 스윕: 경계 지도 (§4.3)
eps 를 훑으며 평균 군집 수를 봅니다. 좁힐수록(작은 eps) 쪼개지고, 넓히면 합의로 모입니다.
세계의 성격이 바뀌는 경계가 어디쯤인지 눈으로 찾아보세요.


```python
for e in (0.05, 0.10, 0.15, 0.20, 0.30):
    d = run_seeds(BoundedConfidence, clusters,
                  seeds=range(1, 21), steps=60, eps=e, n=100)
    print(f'  eps={e:>4}: 평균 군집 {d.mean():.1f} · 범위 {int(d.min())}~{int(d.max())}')
```
